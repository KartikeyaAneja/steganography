```python
from PIL import Image
import numpy as np
```

```python
def getBitStream(imgArr):
    imgArr = imgArr.reshape(-1)
    bitStream = ""
    messageLength = ""
    for i in range(32):
        if imgArr[i] % 2 == 1:
            messageLength += "1"
        else:
            messageLength += "0"
    messageLength = int(messageLength, 2)*8
    for i in range(32, 32+messageLength):
        if imgArr[i] % 2 == 1:
            bitStream += "1"
        else:
            bitStream += "0"
    return bitStream
```

```python
def decrypt(messageBits):
    messageBytes = [messageBits[i:i+8] for i in range(0, len(messageBits), 8)]
    messageBytes = [chr(int(c, 2)) for c in messageBytes]
    return "".join(messageBytes)
```

```python
imgArr = Image.open("encImage.png")
imgArr = np.array(imgArr)
imgArr = imgArr.transpose(2, 0, 1)
messageBits = getBitStream(imgArr)
message = decrypt(messageBits)
print(message)
```
